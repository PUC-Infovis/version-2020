## Preprocesamiento para visualización 2

```python
data_by_artist = pd.read_csv("data_by_artist.csv")
data_by_artist
```

```python
data_by_artist["popularity"].hist()
```

```python
data_by_artist_filtered = data_by_artist.loc[data_by_artist['popularity'] > 60]
data_by_artist_filtered
```

```python
relevant_audio_features = ["valence", "acousticness", "danceability", "energy",
                    "instrumentalness", "liveness", "loudness", "speechiness"]#, "explicit"]
```

```python
artists = data_by_artist_filtered["artists"]
data_by_artist_filtered = data_by_artist_filtered[relevant_audio_features]
data_by_artist_filtered
```

```python
data_by_artist_filtered_norm = (data_by_artist_filtered - data_by_artist_filtered.mean())/data_by_artist_filtered.std()
data_by_artist_filtered_norm
```

```python
sample = data_by_artist_filtered_norm.sample(n = 300)
distances = cdist(sample, sample, 'euclid')
threshold = 0.35 
filtered = np.where(distances <= threshold, distances, 0)
non_zero = np.count_nonzero(filtered)
print(f"Proportion of conections: {non_zero/distances.size}")
print(f"Average connections per artist: {(non_zero/distances.size)*data_by_artist_filtered_norm.shape[0]}")
```

```python
distances = cdist(data_by_artist_filtered_norm, data_by_artist_filtered_norm, 'euclid')
filtered = np.where(distances <= threshold, distances, 0)
```

```python
dict_data = {}
nodes = set()
links = []
for i, distance_row, artist1 in zip(range(len(artists)), filtered, artists):
    if i + 1 <= len(distance_row) - 1:
        for distance, artist2 in zip(distance_row[i+1:], artists[i+1:]):
            if distance != 0:
                link = {"source": artist1, "target": artist2}
                links.append(link)
                if artist1 not in nodes: nodes.add(artist1)
                if artist2 not in nodes: nodes.add(artist2)

dict_data["nodes"] = [{"name": node} for node in nodes]
dict_data["links"] = links
print(f"Amount of nodes: {len(nodes)}")
print(f"Amount of links: {len(links)}")
```

```python
links
```

```python
with open("artists_graph.json", "w") as outfile:  
    json.dump(dict_data, outfile)
```
