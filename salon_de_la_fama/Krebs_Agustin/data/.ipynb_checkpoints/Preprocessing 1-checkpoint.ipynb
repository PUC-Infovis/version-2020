```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial.distance import cdist
import json
```

```python
data = pd.read_csv("data.csv")
data
```

## Preprocesamiento para visualización 1

```python
relevant_audio_features = ["valence", "year", "acousticness", "danceability", "energy", "explicit",
                    "instrumentalness", "liveness", "loudness", "speechiness"]
```

```python
filtered_data = data[relevant_audio_features]
filtered_data
```

## Preprocesamiento para visualización 2

```python
data_by_artist = pd.read_csv("data_by_artist.csv")
data_by_artist
```

```python
data_by_artist.describe()
```

```python
relevant_audio_features = ["valence", "acousticness", "danceability", "energy",
                    "instrumentalness", "liveness", "loudness", "speechiness"]#, "explicit"]
```

```python
artists = data_by_artist["artists"]
data_by_artist = data_by_artist[relevant_audio_features]
data_by_artist
```

```python
data_by_artist = (data_by_artist - data_by_artist.mean())/data_by_artist.std()
data_by_artist
```

```python
sample = data_by_artist.sample(n = 5000)
distances = cdist(sample, sample, 'euclid')
threshold = 0.45 
filtered = np.where(distances <= threshold, distances, 0)
non_zero = np.count_nonzero(filtered)
print(f"Proportion of conections: {non_zero/distances.size}")
print(f"Average connections per artist: {(non_zero/distances.size)*data_by_artist.shape[0]}")
```

```python
distances = cdist(data_by_artist, data_by_artist, 'euclid')
threshold = 0.35 
filtered = np.where(distances <= threshold, distances, 0)
```

```python
dict_data = {}
dict_data["nodos"] = [{"nombre": artist} for artist in artists]
enlaces = []
for i, distance_row, artist1 in zip(range(len(artists)), filtered, artists):
    if i + 1 <= len(distance_row) - 1:
        for distance, artist2 in zip(distance_row[i+1:], artists[i+1:]):
            if distance != 0:
                enlace = {"source": artist1, "target": artist2}
                enlaces.append(enlace)

dict_data["enlaces"] = enlaces
```

```python
with open("artists_graph.json", "w") as outfile:  
    json.dump(dict_data, outfile)
```

```python
def square_to_condensed(i, j, n):
    assert i != j, "no diagonal elements in condensed matrix"
    if i < j:
        i, j = j, i
    return n*j - j*(j+1)//2 + i - 1 - j
```
