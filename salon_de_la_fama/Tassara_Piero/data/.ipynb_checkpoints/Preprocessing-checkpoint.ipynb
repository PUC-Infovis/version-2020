```python
grupos_comunas = comunas.groupby(["COMUNA"])
```

```python
comunas_dict = {}
missing_comunas = ["LA CISTERNA", "SAN RAMÓN", "SAN JOAQUÍN", "CABO DE HORNOS", "SAN MIGUEL"]
# esto es feo pero eficiente :)
done = False
for m in missing_comunas:
    comunas_dict[m.title()] = {}

for c, g in grupos_comunas:
    c = c.title()
    comunas_dict[c] = {}
    g = g.drop(["COMUNA"], axis=1)
    for i in g.index:
        current_period = None
        current_values = {"NOM_COMUNA": c, "PERIOD_DATA": {}}
        
        for h in g:
            if h == "PERIODO":
                current_period = g.loc[i][h]
                if not done:
                    for m in missing_comunas:
                        m = m.title()
                        comunas_dict[m][current_period] = {"NOM_COMUNA": m, "PERIOD_DATA": {}}
            else:
                value = g.loc[i][h]
                if value.dtype == "int64":
                    current_values["PERIOD_DATA"][h] = int(g.loc[i][h])
                elif value.dtype == "float64":
                    current_values["PERIOD_DATA"][h] = float(g.loc[i][h])
                else:
                    current_values["PERIOD_DATA"][h] = int(g.loc[i][h])
                if not done:
                    for m in missing_comunas:
                        m = m.title()
                        comunas_dict[m][current_period]["PERIOD_DATA"][h] = 0
        comunas_dict[c][current_period] = current_values
    done = True
```

```python
for c, v in comunas_dict.items():
    print(c, v)
```

# GeoJson

```python
with open("comunas.geojson", mode="r", encoding="utf8") as file:
    geojson = json.load(file)
```

```python
geojson["features"][0]
```

```python
comuna_id = {}
for f in geojson["features"]:
    comuna_id[f["properties"]["comuna"]] = f["id"]
```

```python
for i in range(len(geojson["features"])):
    c = geojson["features"][i]["properties"]["comuna"]
    c_data = comunas_dict[c]
    geojson["features"][i]["data_CONAF"] = {"NOM_COMUNA": c, "DATA": c_data}
```

```python
geojson["features"][0]
```

```python
with open("comunas_procesadas.geojson", mode="w", encoding="utf8") as file:
    geojson = json.dump(geojson, file, ensure_ascii=False)
```
